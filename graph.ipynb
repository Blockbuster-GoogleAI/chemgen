```python
import networkx as nx
import tkinter as tk
from tkinter import messagebox, scrolledtext

# Define chemical reactions properly 
reactions = {
    ("A", "B"): [("C", "D")],
    ("A", "C"): [("E", "F")],
    ("B", "C"): [("G", None)],
    ("A", "D"): [("H", None), ("X", "Y")], 
    ("C", "D"): [("I", "J")],  
    ("E", "F"): [("K", None)],
    ("I", "J"): [("L", "M")],
}

# Function to recursively build the reaction tree
def build_reaction_tree(reactant1, reactant2, depth=0, visited=None):
    if visited is None:
        visited = set()
    
    if (reactant1, reactant2) in reactions:
        products_list = reactions[(reactant1, reactant2)]
    elif (reactant2, reactant1) in reactions:
        products_list = reactions[(reactant2, reactant1)]
    else:
        return ""

    tree_output = ""
    for products in products_list:
        p1, p2 = products
        tree_output += "    " * depth + f"└─ {reactant1} + {reactant2} → {p1} {('+ ' + p2) if p2 else ''}\n"
        
        # Check further reactions for each product
        for p in products:
            if not p or p in visited:  # Avoid cycles
                continue
            visited.add(p)
            for other in reactions.keys():  # Check if p reacts with other substances
                if p in other:
                    r1, r2 = other
                    tree_output += build_reaction_tree(r1, r2, depth + 1, visited.copy())

    return tree_output

# GUI using Tkinter
def on_submit():
    reactant1 = entry1.get().strip()
    reactant2 = entry2.get().strip()
    reaction_tree = build_reaction_tree(reactant1, reactant2)

    if reaction_tree:
        text_widget.delete("1.0", tk.END)
        text_widget.insert(tk.END, reaction_tree)
    else:
        messagebox.showinfo("Reaction Result", "No reaction path found.")

# Create GUI window
root = tk.Tk()
root.title("Chemical Reaction Finder")

tk.Label(root, text="Reactant 1:").grid(row=0, column=0)
tk.Label(root, text="Reactant 2:").grid(row=1, column=0)

entry1 = tk.Entry(root)
entry2 = tk.Entry(root)
entry1.grid(row=0, column=1)
entry2.grid(row=1, column=1)

submit_button = tk.Button(root, text="Find Reaction", command=on_submit)
submit_button.grid(row=2, column=0, columnspan=2)

# Text area to display reaction tree
text_widget = scrolledtext.ScrolledText(root, width=50, height=20)
text_widget.grid(row=3, column=0, columnspan=2)

root.mainloop()
```
